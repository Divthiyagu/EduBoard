```python
#sets

a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))         # {1, 2, 3, 4, 5}
print(a.intersection(b))  # {3}
print(a)                  # duplicates auto-removed if any
```

```python
#Arithmetic
x = 10
y = 3
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x%y)
print(x//y)
print(x**y)
```

```python
#Comparison
print(10>5)
print(10==10)
print(5!=3)
print(10<5)
print(10!=10)
```

```python
#Logical
a = True
b = False
print(a and b)
print(a or b)
print(not a)
print(not b)
```

```python
#Assignment
x = 5
x += 3
print(x)

y = 5
y -= 2
print(y)

z = 5
z *= 2
print(z)

m = 6
m /= 2
print(m)
```

```python
#strings
s = "hello World"
print(s[0])
print(s[:5])
print(s.upper())
print(s.lower())
print(s.split())
print("-".join(["Hello","World"]))
```

```python
#Lists
fruits = ["apple", "banana", "mango"]
print(fruits[1])          # access
fruits[1] = "orange"      # update
print(fruits)
fruits.append("grapes")   # add
print(fruits)
fruits.extend(["kiwi", "melon"])  # add multiple
print(fruits)
fruits.sort()             # sort list
print(fruits)
fruits.reverse()          # reverse list
print(fruits)
del fruits[0]             # delete item

print(fruits)

```

```python
# Dictionaries
student = {
    "name":"Divya",
    "age":22,
    "Role":"CISCO"
}
print(student["name"])
student["age"]  = 23
print(student["age"])
student["id"] = 101            # add
del student["Role"]            # remove
print(student)

for key,value in student.items():
  print(key,"-",value)
```

```python
#Control Flow

age = 20

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
```

```python
#for loop
for i in range(5):
    print(i)
```

```python
#while loop
x = 1
while x <= 5:
    print(x)
    x += 1
```

```python
#break / continue
for i in range(10):
    if i == 5:
        break   # stop loop
    if i == 2:
        continue # skip 2
    print(i)
```

```python
#break / continue
for i in range(10):
    if i == 2:
        continue # skip 2
    if i == 5:
        break   # stop loop

    print(i)
```

```python
#Functions
def greet():
    print("Hello!")

def add(a, b):
    return a + b

def welcome(name="Guest"):  # default argument
    print("Welcome", name)

greet()
print(add(5, 10))
welcome("Divyasri")
welcome()
```
