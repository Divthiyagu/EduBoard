```python
# Simple Student Report System


def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "Fail"


name = input("Enter your name: ")
age = int(input("Enter your age: "))


subjects = ["Maths", "Science", "English"]


marks = {}

print("\nEnter marks for 3 subjects:")
for s in subjects:
    marks[s] = int(input(f"{s}: "))


total = sum(marks.values())
average = total / 3


grade = get_grade(average)


student = (name, age, grade)


subject_set = set(subjects)


print("\n--- Student Report ---")
print(f"Name: {student[0]}")
print(f"Age : {student[1]}")
print(f"Subjects: {subject_set}")
print(f"Total Marks: {total}")
print(f"Average Marks: {average:.2f}")
print(f"Grade: {student[2]}")


print("\nSubject-wise Marks:")
for sub, m in marks.items():
    print(f"{sub}: {m}")

print("\nThank you for using the Student Report System!")
```

```python
#ascending order [3 9 4 8 1] with out using inbuilt function
a = [3, 9, 4, 8, 1]
for i in range(len(a)):
  for j in range(len(a)-1):
    if(a[j] > a[j+1]):
      temp = a[j]
      a[j] = a[j+1]
      a[j+1] = temp
print(a)
```
